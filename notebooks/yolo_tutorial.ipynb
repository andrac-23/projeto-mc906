```python
import torch
from torchinfo import summary
from IPython.display import Image
```

```python
model = torch.hub.load('ultralytics/yolov5', 'yolov5n', pretrained=True)
summary(model)
```

```python
imgs = ['https://ultralytics.com/images/zidane.jpg']  # batch of images
results = model(imgs)
results.print()
results.save()
```
